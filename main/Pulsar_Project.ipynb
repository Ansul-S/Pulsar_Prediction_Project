```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
```

```python
X = df.drop('Target',axis=1)
y = df['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=101)
```

```python
rbs = RobustScaler()
```

```python
sc_X_train = rbs.fit_transform(X_train)
sc_X_test = rbs.transform(X_test)
```

```python
#Check the shapes
print("Training set shape:", sc_X_train.shape)
print("Test set shape:", sc_X_test.shape)

# Check if scaling worked (values should be around -1 to 1 range mostly)
print("\nBefore scaling - X_train first row:")
print(X_train.iloc[0])
print("\nAfter scaling - sc_X_train first row:")
print(sc_X_train[0])
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=51)
X_train_balanced, y_train_balanced = smote.fit_resample(sc_X_train,y_train)
print("Before SMOTE:")
print(y_train.value_counts())
print("\nAfter SMOTE:")
print(y_train_balanced.value_counts())
```

```python
print(f"Training samples: {X_train_balanced.shape[0]:,}")
print(f"Test samples: {sc_X_test.shape[0]:,}")
print(f"Features: {X_train_balanced.shape[1]}")
print(f"\nClass distribution in training:")
print(f"  Not Pulsar: {(y_train_balanced==0).sum():,} ({(y_train_balanced==0).sum()/len(y_train_balanced)*100:.1f}%)")
print(f"  Pulsar: {(y_train_balanced==1).sum():,} ({(y_train_balanced==1).sum()/len(y_train_balanced)*100:.1f}%)")
print(f"\nClass distribution in test (unchanged):")
print(f"  Not Pulsar: {(y_test==0).sum():,} ({(y_test==0).sum()/len(y_test)*100:.1f}%)")
print(f"  Pulsar: {(y_test==1).sum():,} ({(y_test==1).sum()/len(y_test)*100:.1f}%)")
```

```python
#1st Model: Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

log_model = LogisticRegression(random_state=42, max_iter=1000)
log_model.fit(X_train_balanced,y_train_balanced)

y_pred = log_model.predict(sc_X_test)
```

```python
print(f"\nAccuracy: {accuracy_score(y_test, y_pred):.4f}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
#2nd Model: Random Forest
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100, random_state=101)

rfc.fit(X_train_balanced,y_train_balanced)
y_rfc_predict = rfc.predict(sc_X_test)

print(f"\nAccuracy: {accuracy_score(y_test, y_rfc_predict):.4f}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_rfc_predict))
print("\nClassification Report:")
print(classification_report(y_test, y_rfc_predict, target_names=['Not Pulsar', 'Pulsar']))
```

```python
# Get feature importance
feature_importance = pd.DataFrame({
    'Feature': df.columns[:-1],  # All features except Target
    'Importance': rfc.feature_importances_
}).sort_values('Importance', ascending=False)

print("="*60)
print("FEATURE IMPORTANCE (Random Forest)")
print("="*60)
print(feature_importance)

# Visualize
plt.figure(figsize=(10, 6))
plt.barh(feature_importance['Feature'], feature_importance['Importance'], color='forestgreen')
plt.xlabel('Importance Score', fontsize=12)
plt.ylabel('Features', fontsize=12)
plt.title('Feature Importance in Random Forest', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()  # Highest importance at top
plt.tight_layout()
plt.show()
```

```python
#3rd Model: XGBoost
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=5,
    random_state=101
)

xgb.fit(X_train_balanced,y_train_balanced)

y_xgb_predict = xgb.predict(sc_X_test)


print(f"\nAccuracy: {accuracy_score(y_test, y_xgb_predict):.4f}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_xgb_predict))
print("\nClassification Report:")
print(classification_report(y_test, y_xgb_predict, target_names=['Not Pulsar', 'Pulsar']))
```
